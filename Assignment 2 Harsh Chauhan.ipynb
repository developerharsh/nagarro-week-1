# Qn 1

```python
from math import pi
class circle:
    area=0
    perimeter=0
    def __init__(self,radius):
        self.radius=radius
    def calArea(self):
        self.area=pi*self.radius*self.radius
    def calPeri(self):
        self.perimeter=2*pi*self.radius
        
    def __lt__(self, other): # For x < y
        return self.radius<other.radius

    def __gt__(self, other): # For x > y
        return self.radius>other.radius

    def __le__(self, other): # For x <= y
        return self.radius<=other.radius

    def __ge__(self, other): # For x >= y
        return self.radius>=other.radius

    def __eq__(self, other): # For x == y
        return self.radius==other.radius

    def __ne__(self, other): # For x != y OR x <> y
        return self.radius != other.radius


    
    
```

```python
c1=circle(5)
```

```python
c2=circle(7)
```

```python
c1.calArea()
```

```python
c1.area
```

```python
c1.calPeri()
```

```python
c1.perimeter
```

```python
print(c1<c2)
```

```python
print(c1>c2)
```

```python
print(c1<=c2)
```

```python
print(c1>=c2)
```

```python
print(c1==c2)
```

```python
print(c1!=c2)
```
